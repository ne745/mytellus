# [Tellus の API を学ぶ](https://www.tellusxdp.com/ja/howtouse/dev/20200219_000172.html)

```python
import os
import json

fpth_secret = os.path.join(os.environ['HOME'], 'jupyter/secret.json')
with open(fpth_secret) as f:
    secret = json.load(f)
TOKEN = secret['tellus']
HEADERS = {
    'Authorization': 'Bearer ' + TOKEN
}
```

```python
import requests

params = {
    'min_lat': 32.4440,
    'min_lon': 129.5224,
    'max_lat': 33.4440,
    'max_lon': 130.5224
}

url = 'https://gisapi.tellusxdp.com'
url += '/api/v1/asnaro1/scene'

r = requests.get(url, params=params, headers=HEADERS)
print(json.dumps(r.json(), indent=2))
```

```python
scenes = r.json()
thumbs_url = scenes[-1]['thumbs_url']

from skimage import io
%matplotlib inline
img_thumbs = io.imread(thumbs_url)
io.imshow(img_thumbs)
```
